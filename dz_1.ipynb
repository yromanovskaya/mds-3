# Задание повышенного уровня.

Задан объект в пространстве 2D. 
Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).

```python
import numpy as np
import matplotlib.pyplot as plt

A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
    ]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
A
```

```python
# для этого необходимо к матрице А прибавить вектор c тремя куоординатами: x=300, y=200, z=0
B = A + [300,200,0]
print(B)
```

```python
plt.ylim([-200, 400])
plt.xlim([-200, 500])
x1 = B[:,0]
y1 = B[:,1]
plt.plot(x, y)
plt.plot(x1, y1)
```

Уменьшить объект в два раза.

```python
# для этого необходимо матрицу А умножить на скаляр 1/2
C = A * 1/2
print(C)
```

```python
plt.ylim([-200, 200])
plt.xlim([-200, 200])
x2 = C[:,0]
y2 = C[:,1]
plt.plot(x, y)
plt.plot(x2, y2)
```

Повернуть на 130 градусов.

```python
# для этого необходимо матрицу А умножить на соответсвующую матрицу поворота относительно центра фигуры
from math import cos, sin, pi

theta = 130 * 180/pi
TransformArray_1 = np.array([
    [cos(theta), sin(theta), 0],
    [-sin(theta), cos(theta), 0],
    [0, 0, 1]
])
TransformArray_1
```

```python
D = []
for point in A:
  new_point=np.dot(TransformArray_1,np.transpose(point))
  D.append(new_point)
D = np.array(D).astype(int)
print(D)
```

```python
plt.ylim([-200, 200])
plt.xlim([-200, 200])
x3 = D[:,0]
y3 = D[:,1]
plt.plot(x, y)
plt.plot(x3, y3)
```
